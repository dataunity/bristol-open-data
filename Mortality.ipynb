```python
import pandas as pd
from pandas import DataFrame
import json
import sys
import matplotlib.pyplot as plt
```

```python
# Read the data file and look at some of the data
# Note: you'll need the data file from the following site in a sub directory called 'data'
# http://profiles.bristol.gov.uk/IAS/dataviews/view?viewId=485
df = pd.read_csv('data/dataView485_24.csv')
df['% low birthweight births (<2500g)(2001-2005)']
```

```python
# Output a chart to show results across wards for the latest time period
# Note: create a subdirectory called 'output' before running this code block
col = "Life expectancy at birth(2006-2011)"
expectancy_by_year_cols = ["Life expectancy at birth(2001-2006)",
    "Life expectancy at birth(2002-2007)",
    "Life expectancy at birth(2003-2008)", 
    "Life expectancy at birth(2004-2008)",
    "Life expectancy at birth(2005-2010)",
    "Life expectancy at birth(2006-2011)"]


life_expect = df[["Name", col]]
life_expect = life_expect.set_index("Name")[col]
life_expect_sorted = life_expect.copy()
life_expect_sorted.sort()
plt = life_expect_sorted.plot(kind='bar', 
    figsize=(12, 7))
plt.legend(loc='lower center')
fig = plt.figure
fig.savefig('output/Life-expectancy-at-birth.png')
```

```python
# Output a chart for each ward showing the results over time
wards = [c for c in df['Name'].values]

def create_ward_code(ward):
    # Hack to get rid of undesirable characters in filenames
    ward_code = ward.replace(' ', '-')
    ward_code = ward_code.replace('(', '-')
    ward_code = ward_code.replace(')', '-')
    return ward_code

for ward in wards:
    filtered_by_ward = df[['Name'] + expectancy_by_year_cols][df['Name'] == ward]
    tmp_df = filtered_by_ward.set_index('Name')
    # Truncate column names (ugly hack)
    tmp_df.columns = [c.replace('Life expectancy at birth', '') for c in tmp_df.columns]
    tmp_df
    tmp_df.T.plot().figure.savefig('output/Life-expectancy-at-birth-' + create_ward_code(ward) + '.png')
```

```python
# Quick hack to generate html for the web page
charts_by_ward = ''
wards_sorted = wards
wards_sorted.sort()

for ward in wards_sorted:
    ward_code = create_ward_code(ward)
    print '<a id="top"></a>'
    print '<a href="#%s">%s</a> ' % (ward_code, ward)

for ward in wards_sorted:
    ward_code = create_ward_code(ward)
    print '<h3><a id="%s">%s</a></h3>' % (ward_code, ward)
    print '<div><img src="img/health/Life-expectancy-at-birth-%s.png"/></div>' % ward_code
    print '<div><a href="#top">Back to index</a>'
```

```python
# Generate some JSON code for javascript charts - for all wards
col_name = "Life expectancy at birth(2006-2011)"
pair_data = df[["Name", col_name]]
num_rows = len(pair_data)
output_str = ''
for i in range(num_rows):
    if i > 0:
        output_str += ','
    output_str += '["%s", %s]' % (pair_data['Name'].ix[i], pair_data[col_name].ix[i])
# Hack to get JSON nulls
"[" + output_str.replace('nan', 'null') + "]"
```

```python
# JSON for each ward by year
wards = [c for c in df['Name'].values]
str_output = ''
for ward in wards:
    if str_output != '':
        str_output += ', '
    #ward = 'Ashley'
    filtered_by_ward = df[['Name'] + expectancy_by_year_cols][df['Name'] == ward]
    tmp_df = filtered_by_ward.set_index('Name')
    # Truncate column names
    tmp_df.columns = [c.replace('Life expectancy at birth', '') for c in tmp_df.columns]
    ward_output = ''
    
    for col in tmp_df.columns:
        if ward_output != '':
            ward_output += ','
        ward_output += "[%s, %s]" % (col[1:5], tmp_df.get_value(ward, col))
    str_output += '{"%s": [%s]}' % (ward, ward_output)
str_output = str_output.replace('nan', 'NaN') # Hack to get JSON nulls
str_output
```
